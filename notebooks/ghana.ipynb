Ghana’s “Year of Return” program saw hundreds of African Americans visit the country to experience the history, culture and tradition upfront.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import interpolate
%matplotlib inline

palette = sns.color_palette("bright")
sns.set_context('talk')
sns.palplot(palette)
```

```python
df = pd.read_csv('../data/API_ST.INT.RCPT.CD_DS2_en_csv_v2_718771.csv')
df
```

```python
df.loc[df['Country Name'] == 'Ghana']
```

```python
exports_df = pd.read_csv('../data/API_NE.EXP.GNFS.CD_DS2_en_csv_v2_714634.csv')
exports_df.loc[exports_df['Country Name'] == 'Ghana']
```

```python
arrivals_df = pd.read_csv('../data/API_ST.INT.ARVL_DS2_en_csv_v2_718768.csv')
arrivals_df.loc[exports_df['Country Name'] == 'Ghana']
```

```python
years_1995_to_2017 = list(np.arange(1995, 2018))
tourism_revenue_1995_to_2017 = [ df.loc[df['Country Name'] == 'Ghana'][str(year)].values[0] for year in years_1995_to_2017 ]
exports_1995_to_2017 = [ exports_df.loc[exports_df['Country Name'] == 'Ghana'][str(year)].values[0] for year in years_1995_to_2017 ]
years_1995_to_2019 = list(np.arange(1995, 2020))
tourism_revenue_1995_to_2017_zeros_to_2019 = tourism_revenue_1995_to_2017.copy()
tourism_revenue_1995_to_2017_zeros_to_2019.extend([0, 0])
tourism_revenue_2019 = [ 0 for year in years_1995_to_2019 ]
tourism_revenue_2019[-1] = 1900000000
np.array(tourism_revenue_1995_to_2017)/np.array(exports_1995_to_2017)
```

```python
%run ../python_files/visualizations
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
sns.barplot(years_1995_to_2019, tourism_revenue_1995_to_2017_zeros_to_2019, color=palette[9], label='1995-2017 (Actual)', ax=ax);
sns.barplot(years_1995_to_2019, tourism_revenue_2019, color=palette[6], label='2019 (Estimated)', ax=ax);
plt.title('Ghana generated $1.9 Billion through activities related to the "Year of Return"')
plt.ylabel('International Tourism Receipts (Billions of $ USD)');
plt.xticks(rotation=90);
plt.legend()
sns.despine()
plt.tight_layout()
```

Ghana’s economy generated a total of $1.9 billion through activities related to the “Year of Return”, according to Tourism Minister Barbara Oteng-Gyasi.

She said: “There has been tremendous community involvement which has stimulated the local economy including hoteliers, tour operating and other related businesses.”

```python
tourism_revenue_1995_to_2017_zeros_to_2027 = tourism_revenue_1995_to_2017_zeros_to_2019.copy()
tourism_revenue_2018_zeros_to_2027 = tourism_revenue_2018.copy()
tourism_revenue_2019_zeros_to_2027 = tourism_revenue_2019.copy()
tourism_revenue_1995_to_2017_zeros_to_2027.extend([0,0,0,0,0,0,0,0])
tourism_revenue_2018_zeros_to_2027.extend([0,0,0,0,0,0,0,0])
tourism_revenue_2019_zeros_to_2027.extend([0,0,0,0,0,0,0,0])

years_1995_to_2027 = list(range(1995, 2028))
years_2019_to_2027 = list(range(2019, 2028))
tourism_revenue_2027 = [ 0 for year in years_1995_to_2027 ]
tourism_revenue_2019_to_2027 = np.linspace(1900000000, 8375500000, num=9, endpoint=True)
tourism_revenue_2020_to_2027 = tourism_revenue_2019_to_2027[1:]
years_2020_to_2027 = years_2019_to_2027[1:]
```

The country recently unveiled a 15-year-long tourism plan that seeks to increase the annual number of tourists to Ghana from one million to eight million per year by 2027.

Ghana's travel industry is projected to raise $8.3 billion a year by 2027, plus associated benefits, according to the plan.

```python
offset = len(tourism_revenue_2019)

for i, rev in enumerate(tourism_revenue_2020_to_2027):
    tourism_revenue_2027[i + offset] = tourism_revenue_2020_to_2027[i]
    
```

```python
fig, ax = plt.subplots(figsize=(16, 10))
sns.barplot(years_1995_to_2027, tourism_revenue_1995_to_2017_zeros_to_2027, color=palette[9], label='1995-2017 (Actual)', ax=ax);
sns.barplot(years_1995_to_2027, tourism_revenue_2018_zeros_to_2027, color=palette[7], label='2018 (Imputed)', ax=ax);
sns.barplot(years_1995_to_2027, tourism_revenue_2019_zeros_to_2027, color=palette[6], label='2019 (Estimated)', ax=ax);
sns.barplot(years_1995_to_2027, tourism_revenue_2027, color=palette[4], label='2020-2027 (Projected)', ax=ax);
plt.title("Ghana's tourism plan projects $8.3 billion annual receipts by 2027")
plt.ylabel('International Tourism Receipts (Billions of $ USD)');
plt.xticks(rotation=90);
plt.legend()
sns.despine()
plt.tight_layout()
```

According to data from ‘The Year of Return’ Secretariat in the first nine months of 2019, there have been 80,862 more arrivals from the United States of America, the United Kingdom, and other key target nations than the figure recorded the previous year.

The figures showed that in-bound airport arrivals from the United States of America increased by 26 percent (17,455) and the United Kingdom by 24 percent (9,812). An official noted, “Total airport arrivals have increased by 45 percent (237,088), compared to last year.”
